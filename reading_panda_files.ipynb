** filtering with loc function **

```python
df1.loc[df1.A < 5] # selects only rows where column A value is less than 5
```

```python
df1.loc[df1['C'] == 1.4] # this will look at the df1 of C if the condition of equal to 1.4 (see below table) is true, it will print out the whole row
```

```python
df1['C'] == 1.4
```

```python
df1.loc[df1['C'] == 1.4, 'Label'] #gives just the labels for those rows
```

```python
df1['D'].loc[lambda x: x > 2.1] # prints only rows of data where rows is greater than 2.1
```

```python
df1.iloc[:10]
```

```python
df1.index = [i for i in range (0,750,5)] # counts numerically from the beginning irrespective of whar the index is
df1.iloc[:10]
```

```python
df1.iloc[:, 1].head(6)
```

```python
df1.iloc[1:7, 2]
```

```python
df1.query('B > C & C < 4*D') #querry uses parenthesis and takes in boolean. Always have  a quotation mark around the querry if you are using boolean
```

```python
df1.query('A in [5.7,5.9]') # produces all rows that have either 5.7 or 5.9 for A
```

```python
df1.query('A != [5.7,5.9]')
```

```python
df1.query('B > 3.5 & D != [0.2, 1.0, 1.3, 1.4, 1.5, 1.8, 2.0, 2.3]') #query can handle many options
```

```python
df2['20131001': '2015'] # python knows how to convert tge format into date and prints date from 20131001 till 2015
```
